# Reactor Reactivity experiments

```python
from matplotlib import pyplot as plt

def reactivity():
    fastFissionP = 1.03
    fastNonLeakageP = 0.95
    resonanceEscapeP = 0.75
    thermalNonLeakageP = 0.96
    thermalUtilizationP = 0.70 # 0.699222
    reproductionFactor = 2.03
    kEff = fastFissionP * fastNonLeakageP * resonanceEscapeP * thermalNonLeakageP * thermalUtilizationP * reproductionFactor
    reactivity = (kEff - 1) / kEff
    return kEff, reactivity

def neutron_flux(initial, seconds):
    runs = seconds / 0.1
    factor = (reactivity()[0]) ** runs
    return initial * factor

initial_flux = 1
seconds = 1000
time = [x for x in range(seconds)]
flux = [neutron_flux(initial_flux, x) for x in time] 

plt.plot(time, flux)
plt.ylim(bottom=0)
plt.xlim(left=0)
plt.ticklabel_format(style="plain")
plt.show()
```
